### In this excersize, generate some basic charts using Pandas. 


```python
# import data from pandas
import pandas as pd
import os

#Make directory called plots
os.makedirs("plots", exist_ok=True)

#create the DataFrame utilizing Pandas 
data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv") 

#generate all info for this dataset
data.info()

#Remove the last line of NAN
data = data.iloc[:-1 , :]

#Ignore any warnings
import warnings
warnings.filterwarnings('ignore')
```

###  Part One: In this excersize, generate a basic histogram.


```python
#Please generate a basic histogram and save the histogram as a png file
# Hint: use pandas histogram function.

# YOUR SOLUTION HERE 
hist = data.plot.hist()
histpng = hist.figure.savefig('plots/hist.png')



# ANSWER CHECK
print(hist)

# ax = s.plot.hist()
# ax.figure.savefig('demo-file.pdf')
```

###  Part Two: In this excersize, generate a basic scatter plot. 


```python
#Please generate a basic scatter plot utilizing Age as the x axis and Salary as the y axis
# Hint: use pandas scatter function to generate the chart.

# YOUR SOLUTION HERE 
scat = data.plot.scatter(x='Age',y='Salary')

scatpng = scat.figure.savefig('plots/scat.png')


# ANSWER CHECK
print(scat)


```

###  Part Three: In this excersize, generate a basic lineplot. 

```python
# Please generate a basic line plot by name as the x axis and salary as the y axis 
#for the "Golden State Warriors" using the nba data set. 
#Note a new dataframe called teamdata has been created for you that will be used during the rest of this 

teamdata = data.filter(items=['Team', 'Name', 'Salary'])
teamdata = teamdata.loc[data.Team == 'Golden State Warriors', 'Name':'Salary']
teamdata.set_index('Name')

# YOUR SOLUTION HERE 
lines = teamdata.plot.line()


lines.set_xticklabels(teamdata.Name, rotation=90) 
linepng = lines.figure.savefig('plots/linepng.png')

# ANSWER CHECK
print(lines)
```

###  Part Four: In this excersize, generate a basic bar chart. 

```python
# Please generate a basic scatter plot by name as the x axis and salary as the y axis of the teamdata dataframe.
 

# YOUR SOLUTION HERE 
bar = teamdata.plot.bar(x="Name", y="Salary")

barpng = bar.figure.savefig('plots/barpng.png')

# ANSWER CHECK
print(bar)

```
