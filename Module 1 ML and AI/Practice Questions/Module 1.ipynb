```python
#With the DataFrame df= pd.DataFrame({’A’:[0.1,0.2,0.3], ‘B’:[10,20,30], ‘C’:[1,2,3]}), how would you create a histogram that includes columns B and C?#
df= pd.DataFrame({'A':[0.1,0.2,0.4], 'B':[10,20,8], 'C':[1,2,8]})
df.plot(kind = 'hist',y=['B','C'])
```
